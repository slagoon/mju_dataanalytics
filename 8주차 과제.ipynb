```python
import csv

with open('./data/homework2.csv', 'w', encoding = 'utf-8-sig', newline='') as writer_csv:
    name_list = ['phone','price']
    writer = csv.DictWriter(writer_csv, fieldnames = name_list)
    writer.writeheader()
    writer.writerow({'phone' : 'iphone12pro', 'price' : '1350000'})
    writer.writerow({'phone' : 'iphone12pro max', 'price' : '1490000'})
    writer.writerow({'phone' : 'iphone12', 'price' : '1090000'})
    writer.writerow({'phone' : 'iphone12 mini', 'price' : '950000'})
    
```
