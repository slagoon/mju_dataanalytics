```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
#변수의 상관관계 확인
titanic.corr()
```

- 숫자가 클수록 연관성이 크다

```python
#변수간 상관관계 히트맵
sns.heatmap(titanic.corr(), annot = True, cmap = 'GnBu', linewidths = 0.2)
fig = plt.gcf()
fig.set_size_inches(10,8)
plt.show()
```

```python
#원핫인코딩 - 더미변수만들기
pd.get_dummies(titanic['embark_town'])
```

```python
df1 = pd.get_dummies(data = titanic, columns = ['embark_town'], prefix = 'E')
```

```python
df1
```

```python
sns.heatmap(df1.corr(), annot = True, cmap = 'Blues_r', linewidths = 0.2)
fig = plt.gcf()
fig.set_size_inches(10,8)
plt.show()
```
