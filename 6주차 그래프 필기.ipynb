# 여러개 그래프 그리기

```python
# 축 제목 지정
plt.plot(t,t, 'r--', t, t**2,'bs', t, t**3, 'g*')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([0,5,0,20])
```

```python
plt.plot([1,2,3,4],[1,4,9,16], color = 'springgreen', marker = '*', linestyle = '--')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([0,5,0,20])
```

```python
# 그래프 영역 채우기
x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.fill_between(x[1:3], y[1:3], alpha = 0.5) # X축에 색을 칠하는 방법
#plt.fill_betweenx(y[1:3], x[1:3], color = 'springgreen', alpha = 0.5)
plt.show()
```

```python
# 그래프 영역 채우기
x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')

plt.fill_betweenx(y[1:3], x[1:3], color = 'springgreen', alpha = 0.5)
plt.show()
```

```python
# 2개의 그래프 사이에 영역 채우기
x = [1,2,3,4]
y1 = [1,4,9,16]
y2 = [1,2,4,8]

plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('X')
plt.ylabel('Y')
plt.fill_between(x[1:3],y1[1:3],y2[1:3],alpha = 0.5)

plt.show()
```

# 그리드 설정하기

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)

plt.grid(True, axis='x') # x축만 격자 생성

plt.show()
```

# 눈금 표시하기

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)

plt.grid(color='red', alpha=0.4, linestyle='--') # 그리드 색상, 라인스타일 지정 가능

# 틱 지정 (간격, 레이블)
plt.xticks(a, labels=['Jan','','Feb','','Mar','','May','','June',''])
plt.yticks(np.arange(0,7), ('0','1GB','2GB','3GB','4GB','5GB','6GB'))

plt.show()
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)

# 틱 지정 (간격, 레이블)
plt.xticks(a, labels=['Jan','','Feb','','Mar','','May','','June',''])
plt.yticks(np.arange(0,7), ('0','1GB','2GB','3GB','4GB','5GB','6GB'))

# 틱 파라미터 조정
# X축, 눈금 안쪽 방향, 눈금길이, 레이블 여백, 레이블 크기, 레이블 색상
plt.tick_params(axis='x', direction='in', length=10, pad=6, labelsize=13, labelcolor='green')

# Y축, 눈금 안쪽 방향, 눈금길이, 레이블 여백, 레이블 크기, 레이블 색상
plt.tick_params(axis='y', direction='in', length=20, pad=20, labelsize=13, labelcolor='orange')

plt.show()
```

```python
np.arange(0,2,0.2)

plt.plot(a,a, 'bo')
plt.plot(a, a**2, color = 'yellow', marker = '*', linewidth = 2)
plt.plot(a, a**3, color = 'springgreen', marker = '^', markersize = 9)

plt.xticks(np.arange(0,2,0.2), labels = ['jan','','Feb','','Mar','','May','','June','',])
plt.yticks(np.arange(0,7), ('0','1GB','2GB','3GB','4GB','5GB','6GB'))

plt.tick_params(axis = 'x', direction = 'inout', length=3, pad=6, labelsize = 15, labelcolor = 'springgreen')


#pad는 눈금과 레이블의 거리,direction은 눈금의 방향을 지정
plt.tick_params(axis = 'y', direction = 'inout', length=3, pad=6, labelsize = 15, labelcolor = 'springgreen')
plt.show()
```

# 타이틀 지정하기

```python
a = np.arange(0,2,0.2)

plt.plot(a,a, 'bo')
plt.plot(a, a**2, color = 'black', marker = '*', linewidth = 2)
plt.plot(a, a**3, color = 'springgreen', marker = '^', markersize = 9)
plt.grid(True, axis = 'y', color= 'gray', alpha = 0.4, linestyle = '--')
plt.tick_params(axis = 'both', direction = 'inout', length= 3, pad = 6, labelsize = 14)
plt.title('Sample', loc = 'right', pad=25
         )
title_font = {
    'fontsize' : 16,
    'fontweight' : 'bold'
}
plt.title('Sample', fontdict = title_font, loc = 'left', pad = 25)

plt.show()
```

# 막대 그래프 그리기

```python
y = np.arange(3)

years = ['2017','2018','2019']
values = [100,400,900]

plt.bar(y, values, width = 0.4, align = 'edge', color = 'green',edgecolor = 'black')
plt.xticks(y, years)
plt.show()
```

```python
y = np.arange(3)

years = ['2017','2018','2019']
values = [100,400,900]

plt.barh(y, values, align = 'edge', color = 'green',edgecolor = 'black',
       linewidth = 3, tick_label = years, log = False)

plt.show()
```

# 산점도 그리기

```python
np.random.seed(123)#난수 발생을 위해 써주는 값 0보다 큰값만 너어주면 됨

N=50
x = np.random.rand(N)#50개의 x좌표
y = np.random.rand(N)#50개의 y좌표
colors = np.random.rand(N)
area = (30*np.random.rand(N))**2

plt.scatter(x,y, s=area, c=colors, alpha = 0.4)
plt.show()
```

# 파이차트 그리기

```python
ratio = [34,32,16,18]
labels = ['apple','banana','melon','grapes']

plt.pie(ratio, labels = labels, autopct = '%.1f%%')# autopct 소수점 자리수
plt.show()
```

```python
ratio = [34,32,16,18]
labels = ['apple','banana','melon','grapes']
colors = ['#ff9999', '#ffc000','#8fd9b6','#d395d0']
wedgeprops = {'width' : 0.7, 'edgecolor':'w', 'linewidth':5}

plt.pie(ratio, labels = labels, autopct = '%.1f%%', startangle=260, counterclock=False,colors=colors, wedgeprops =wedgeprops)
plt.show()
```
