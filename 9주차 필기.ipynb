# API를 활용하여 데이터 추출하기

```python
from bs4 import BeautifulSoup
import requests

keyword = '태연'
api = 'http://suggestqueries.google.com/complete/search?output=toolbar&q=' + keyword

response = requests.get(api)
soup = BeautifulSoup(response.content,'xml')
datas=soup.select('suggestion')

for item in datas:
    print(item['data'])
```

# JSON(JavaScript Object Notation)
 - 서버와 클라이언트 또는 컴퓨터/프로그램사이에 데이터를 주고 받을 때 사용하는 데이터 형식

### Json 데이터 포맷 일기
- josn 라이브러리 제공
 - json.loads()함수로 문자열로된 json 데이터를 사전처럼 다룰 수 있음
 

```python
import json
```

```python
#문자열로 된 데이터를 입력 받아 json 포맷으로 변환
data = '{"id":"1001","language":"Python","title":"Programming for Data Analytics"}'
jsondata = json.loads(data)
```

```python
jsondata['id']
```

```python
jsondata['language']
```

```python
type(jsondata)
```

```python
#딕셔너리로 된 데이터를 입력 받아 json 포맷으로 변환
data = {"id":"1001","language":"Python","title":"Programming for Data Analytics"}
type(data)
```

```python
jsondata = json.dumps(data)
```

```python
jsondata
```

```python
type(jsondata)
```

```python
data = {"id":"1001","language":{"Python":"basic","Java":"advance"},"title":"Programming for Data Analytics"}
```

```python
jsondata = json.dumps(data,indent = 2)
```
